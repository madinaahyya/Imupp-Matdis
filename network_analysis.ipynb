````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# --- Pilih variabel untuk radar chart ---
coding_cols = ['X5','X6','X7','X8']
ai_cols = ['X9','X10','X11','X12']
vars_all = coding_cols + ai_cols

# --- Hitung rata-rata tiap cluster ---
cluster_mean = df.groupby('Cluster')[vars_all].mean()

# --- Radar Chart ---
labels = vars_all
num_vars = len(labels)

angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # ulangi titik pertama untuk menutup radar

plt.figure(figsize=(8,8))

for cluster_id in cluster_mean.index:
    values = cluster_mean.loc[cluster_id].tolist()
    values += values[:1]  # tutup grafik

    plt.polar(angles, values, marker="o", label=f"Cluster {cluster_id}")

# Styling
plt.xticks(angles[:-1], labels, fontsize=10)
plt.title("Radar Chart – Coding vs AI Behavior per Cluster", size=14)
plt.legend(loc="upper right", bbox_to_anchor=(1.2, 1.1))
plt.grid(True)

plt.show()
````

**sequence analysis**

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- Load Data ---
df = df.copy()

# --- Ambil kolom sequence ---
sequences = df[['X13', 'X14']].fillna("Missing")

# --- Bersihkan teks sequence menjadi list step-step ---
def clean_seq(text):
    steps = [s.strip() for s in text.split("→")]
    return steps

df['seq1'] = df['X13'].apply(clean_seq)
df['seq2'] = df['X14'].apply(lambda x: [x])

# Gabungkan urutan aktivitas utama
df['full_sequence'] = df.apply(lambda x: x['seq1'] + x['seq2'], axis=1)

# --- Bangun Transition Matrix ---
from collections import Counter, defaultdict

transitions = defaultdict(Counter)

for seq in df['full_sequence']:
    for i in range(len(seq)-1):
        transitions[seq[i]][seq[i+1]] += 1

# Convert ke dataframe
states = sorted(
    set([s for t in transitions for s in transitions[t]]) |
    set([n for t in transitions for n in transitions[t].keys()])
)

matrix = pd.DataFrame(0, index=states, columns=states)

for a in transitions:
    for b in transitions[a]:
        matrix.loc[a, b] = transitions[a][b]

# --- Visualisasi Transition Heatmap ---
plt.figure(figsize=(12,8))
sns.heatmap(matrix, annot=False, cmap="Purples", linewidths=.5)
plt.title("Transition Matrix — Coding → Next Activities (Markov)")
plt.show()
````

**network analysis**

````python
import networkx as nx

# --- Build Graph ---
G = nx.Graph()

# Tambahkan node coding
coding = ['X5','X6','X7','X8']
ai = ['X9','X10','X11','X12']

G.add_nodes_from(coding, group='coding')
G.add_nodes_from(ai, group='ai')

# Tambahkan edges berdasarkan korelasi kuat
corr = df[coding + ai].corr(method='spearman')

threshold = 0.3  # kamu bisa naikkan untuk lebih clean

for a in corr.columns:
    for b in corr.columns:
        if a!=b and abs(corr.loc[a,b]) > threshold:
            G.add_edge(a,b, weight=float(abs(corr.loc[a,b])))

# --- Centrality ---
bet = nx.betweenness_centrality(G)
deg = nx.degree_centrality(G)

print("\n=== Betweenness Centrality ===")
for k,v in bet.items():
    print(k,":", round(v,4))

print("\n=== Degree Centrality ===")
for k,v in deg.items():
    print(k,":", round(v,4))

# --- Community Detection (Louvain) ---
from networkx.algorithms.community import greedy_modularity_communities

communities = list(greedy_modularity_communities(G))

print("\n=== COMMUNITIES FOUND ===")
for i, comm in enumerate(communities):
    print(f"Community {i+1}: {list(comm)}")

# --- Visualisasi ---
plt.figure(figsize=(8,6))
pos = nx.spring_layout(G, k=0.5)

nx.draw(G, pos,
        with_labels=True,
        node_color=['lightblue' if G.nodes[n]['group']=='coding' else 'pink' for n in G.nodes()],
        node_size=1400,
        font_size=10,
        edge_color="gray")

plt.title("Graph: Coding vs AI Activities")
plt.show()
````

````python
import numpy as np
import pandas as pd
import networkx as nx
import plotly.graph_objects as go
from sklearn.metrics.pairwise import cosine_similarity

# -----------------------------
# BASIC CLEANING: map categorical frequency/status if present
# -----------------------------
map_status = {"Pemula": 1, "Menengah": 2}
map_jenis = {"Proyek pribadi": 1, "Tugas Kuliah": 2}
map_waktu = {"30 menit - 1 jam": 1, "> 2 jam": 2}
map_freq = {"≤ 5 Kali": 1, "> 10 Kali": 2}

for c in ["X1","X2","X3","X4"]:
    if c in df.columns:
        if c == "X1":
            df[c] = df[c].map(map_status)
        elif c == "X2":
            df[c] = df[c].map(map_jenis)
        elif c == "X3":
            df[c] = df[c].map(map_waktu)
        elif c == "X4":
            df[c] = df[c].map(map_freq)

# -----------------------------
# SELECT numeric columns used for similarity (X5..X12)
# -----------------------------
numeric_cols = ["X5","X6","X7","X8","X9","X10","X11","X12"]
# Keep only columns that exist in df
numeric_cols = [c for c in numeric_cols if c in df.columns]

if len(numeric_cols) == 0:
    raise ValueError("Tidak ada kolom numeric X5..X12 ditemukan di dataset. Pastikan kolom ada.")

# Force numeric and fill NaN with column median (safer)
df_numeric = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
df_numeric = df_numeric.fillna(df_numeric.median())

names = df["Nama"].astype(str).tolist()
n = len(names)
print(f"Using {len(numeric_cols)} numeric columns for similarity. Number of persons = {n}.")

# -----------------------------
# SIMILARITY MATRIX (cosine)
# -----------------------------
sim = cosine_similarity(df_numeric.values)

# show basic stats
tri_idx = np.triu_indices_from(sim, k=1)
sim_values = sim[tri_idx]
print(f"Similarity stats (pairwise, upper-triangle): mean={sim_values.mean():.4f}, std={sim_values.std():.4f}, min={sim_values.min():.4f}, max={sim_values.max():.4f}")

# -----------------------------
# ADAPTIVE THRESHOLD (mean + std) — ensures graph not full & not empty
# -----------------------------
if sim_values.size == 0:
    threshold = 0.9
else:
    threshold = sim_values.mean() + sim_values.std()

# clamp threshold to [0.5, 0.98] for sanity (adjustable)
threshold = float(np.clip(threshold, 0.5, 0.98))
print(f"Adaptive threshold used: {threshold:.4f}")

# -----------------------------
# BUILD GRAPH based on similarity threshold
# -----------------------------
G = nx.Graph()
G.add_nodes_from(names)

for i in range(n):
    for j in range(i+1, n):
        if sim[i, j] >= threshold:
            G.add_edge(names[i], names[j], weight=float(sim[i, j]))

print(f"Graph: nodes={G.number_of_nodes()}, edges={G.number_of_edges()} (threshold {threshold:.4f})")

# If graph has 0 edges, relax threshold fallback
if G.number_of_edges() == 0:
    print("Graph has 0 edges with adaptive threshold — relaxing threshold to mean(sim) and then 0.5 fallback.")
    thr2 = sim_values.mean() if sim_values.size>0 else 0.5
    thr2 = float(np.clip(thr2, 0.2, 0.95))
    print("Fallback threshold:", thr2)
    for i in range(n):
        for j in range(i+1, n):
            if sim[i, j] >= thr2:
                G.add_edge(names[i], names[j], weight=float(sim[i, j]))
    print(f"After fallback: edges={G.number_of_edges()}")
    if G.number_of_edges() == 0:
        # final fallback: connect top-k most similar pairs (k = n)
        print("Still 0 edges — connecting top-n pairs to avoid empty graph.")
        pairs = []
        for i in range(n):
            for j in range(i+1, n):
                pairs.append((sim[i,j], i, j))
        pairs = sorted(pairs, key=lambda x: x[0], reverse=True)
        k = max(1, n)  # connect at least n edges
        for t in pairs[:k]:
            _, i, j = t
            if not G.has_edge(names[i], names[j]):
                G.add_edge(names[i], names[j], weight=float(sim[i,j]))
        print(f"After top-k connect: edges={G.number_of_edges()}")

# -----------------------------
# BRIDGE SCORES (betweenness centrality)
# -----------------------------
bridge_scores = nx.betweenness_centrality(G, weight='weight', normalized=True)
# Sort descending
bridge_sorted = sorted(bridge_scores.items(), key=lambda x: x[1], reverse=True)
print("\n=== Bridge Person Scores (top -> bottom) ===")
for name, score in bridge_sorted:
    print(f"{name:35s} : {score:.6f}")

# Save bridge scores to CSV
bridge_df = pd.DataFrame(bridge_sorted, columns=['Nama','BridgeScore'])
bridge_df.to_csv("bridge_scores.csv", index=False)
print("Saved bridge_scores.csv")

# -----------------------------
# PLOTLY INTERACTIVE GRAPH and SAVE HTML
# -----------------------------
pos = nx.spring_layout(G, seed=42, k=None)  # for larger graphs consider kamada_kawai_layout

# Edges traces
edge_x = []
edge_y = []
edge_weights = []
for u, v, d in G.edges(data=True):
    x0, y0 = pos[u]
    x1, y1 = pos[v]
    edge_x += [x0, x1, None]
    edge_y += [y0, y1, None]
    edge_weights.append(d.get('weight', 1.0))

edge_trace = go.Scatter(
    x=edge_x,
    y=edge_y,
    line=dict(width=0.5, color='rgba(150,150,150,0.4)'),
    hoverinfo='none',
    mode='lines'
)

# Node traces
node_x = []
node_y = []
node_text = []
node_size = []
node_color = []

# normalize bridge scores for marker sizing/color
if len(bridge_scores) > 0:
    scores_arr = np.array(list(bridge_scores.values()))
    max_s = scores_arr.max() if scores_arr.max() > 0 else 1.0
else:
    max_s = 1.0

for node in G.nodes():
    x, y = pos[node]
    node_x.append(x); node_y.append(y)
    sc = bridge_scores.get(node, 0.0)
    node_text.append(f"{node}<br>Bridge Score: {sc:.6f}")
    # size: base + scaled
    node_size.append(8 + (sc / max_s) * 25)
    node_color.append(sc)

node_trace = go.Scatter(
    x=node_x,
    y=node_y,
    mode='markers',
    hoverinfo='text',
    text=node_text,
    marker=dict(
        showscale=True,
        colorscale='Inferno',
        reversescale=False,
        color=node_color,
        size=node_size,
        colorbar=dict(title="Bridge Score"),
        line_width=0.5
    )
)

fig = go.Figure(data=[edge_trace, node_trace],
                layout=go.Layout(
                    title=f'Interactive Bridge Network (n={n}, threshold={threshold:.3f})',
                    title_x=0.5,
                    showlegend=False,
                    hovermode='closest',
                    margin=dict(b=20,l=5,r=5,t=40),
                    annotations=[ dict(
                        text="Graph based on cosine similarity of numeric activity features (X5..X12).",
                        showarrow=False,
                        xref="paper", yref="paper",
                        x=0.005, y=-0.002 ) ],
                    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False)
                )
)

out_html = "bridge_network.html"
fig.write_html(out_html, include_plotlyjs='cdn')
print(f"Saved interactive HTML to {out_html}  — buka file ini di browser.")
````
