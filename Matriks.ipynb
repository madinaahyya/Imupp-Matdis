```python
import pandas as pd
import numpy as np
data = pd.read_csv('/content/datamatdis_imupp.csv')
data
```

```python
data.dtypes
```

```python
cols = ['X1','X2','X3','X4']

# IDEA 1: Unweighted adjacency by exact match on at least one selected attribute
n = len(data)
adj1 = np.zeros((n, n), dtype=int)

for i in range(n):
    for j in range(n):
        if i != j:
            if any(data.loc[i, c] == data.loc[j, c] for c in cols):
                adj1[i, j] = 1

# IDEA 2: Weighted adjacency (similarity count)
adj2 = np.zeros((n, n), dtype=int)
for i in range(n):
    for j in range(n):
        if i != j:
            adj2[i, j] = sum(data.loc[i, c] == data.loc[j, c] for c in cols)

# IDEA 3: Matrix of hours categories (X3)
cats = data['X3'].unique()
cat_index = {c:i for i,c in enumerate(cats)}
k = len(cats)
mat3 = np.zeros((k, k), dtype=int)

# Count members in each category → diagonal matrix
for c in data['X3']:
    idx = cat_index[c]
    mat3[idx, idx] += 1

adj1[:5,:5], adj2[:5,:5], mat3
```
